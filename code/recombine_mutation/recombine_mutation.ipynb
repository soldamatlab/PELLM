Recombine mutation

```python
%load_ext autoreload
%autoreload 2

from recombine_mutation import main
import random
```

```python
random.seed(10)
best_variant, fitness_progression = main()
```

Save / Load results

```python
import pickle
file_name = 'results.pkl'
```

```python
with open(file_name, 'wb') as f:
    pickle.dump([best_variant, fitness_progression], f)
```

```python
with open(file_name, 'rb') as f:
    best_variant, fitness_progression = pickle.load(f)
```

Plot learning curve

```python
import matplotlib.pyplot as plt
```

```python
n_evaluations = [i for i in range(1, len(fitness_progression)+1)]
plt.plot(n_evaluations, fitness_progression)
plt.xlabel('Number of Evaluations')
plt.ylabel('Highest Fitness Reached (relative to WT variant)')
finish = 'Final variant: ' + best_variant + ' ({:.2f})'.format(fitness_progression[-1])
plt.title('Recombine mutation - GB1\n' + finish)
plt.show()
```
