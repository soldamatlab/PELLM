```python
%load_ext autoreload
%autoreload 2

import torch
from two_layer_perceptron import TwoLayerPerceptron
```

```python
embedding_size = 1280

model = TwoLayerPerceptron(torch.nn.Sigmoid(), embedding_size)
model.eval()
```

```python
out = torch.zeros((10000, 1))
with torch.no_grad():
    for i in range(len(out)):
        input = torch.rand((1, embedding_size))
        out[i] = model(input)
```

```python
import matplotlib.pyplot as plt

plt.hist(out.squeeze(), bins=10)
plt.show()
```
